### Model Building

#### Defining Function for all types of Machine Learning Algorithm

```python
def Multiple_Linear_Regression(df):
    x = df.drop('Adj Close',axis=1)
    y = df['Adj Close']
    # Splitting data
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    
    LR = LinearRegression()
    LR.fit(x_train,y_train)
    
    y_prediction = LR.predict(x_test)
    return r2_score(y_test,y_prediction)
```

#### Multiple Linear Regression

```python
MLR_accuracy_df = pd.DataFrame({'DataFrame':[],'Accuracy':[]}) 
for key,value in df_stock_dict.items():
    print("{} -> {}".format(key,Multiple_Linear_Regression(df_stock_dict[key])))
    MLR_accuracy_df.loc[len(MLR_accuracy_df.index)] = [key,Multiple_Linear_Regression(df_stock_dict[key])]
```

```python
MLR_accuracy_df = MLR_accuracy_df.sort_values(by = 'Accuracy')
```

```python
#define plot size
plt.figure(figsize=(10,5))

plt.plot(MLR_accuracy_df['DataFrame'],MLR_accuracy_df['Accuracy'],label='Multiple Linear Regression')
plt.xlabel('Data-Frames')
plt.ylabel('Accuracy')
plt.show()
```

#### Looking for Data-frame's models that performed poor

```python
for row in MLR_accuracy_df.index:
    if MLR_accuracy_df['Accuracy'][row] < 0.8:
        print(MLR_accuracy_df['DataFrame'][row],end=",")
```

### Applying Boosting Algorithms

#### Gradient Boosting Algorithm

```python
def Gradient_Boosting(df):
    x = df.drop('Adj Close',axis=1)
    y = df['Adj Close']
    # Splitting data
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    
    #transforming target variable through quantile transformer
    regr_trans = TransformedTargetRegressor(regressor=GradientBoostingRegressor(), transformer=QuantileTransformer(output_distribution='normal'))
    regr_trans.fit(x_train, y_train)
    yhat = regr_trans.predict(x_test)
    algoname= GradientBoostingRegressor().__class__.__name__
    return round(r2_score(y_test, yhat),3)
```

```python
GB_accuracy_df = pd.DataFrame({'DataFrame':[],'Accuracy':[]}) 
for key,value in df_stock_dict.items():
    print("{} -> {}".format(key,Gradient_Boosting(df_stock_dict[key])))
    GB_accuracy_df.loc[len(GB_accuracy_df.index)] = [key,Gradient_Boosting(df_stock_dict[key])]
```

##### Visualizing improvements

```python
#define plot size
plt.figure(figsize=(20,8))

plt.plot(MLR_accuracy_df['DataFrame'],MLR_accuracy_df['Accuracy'],label='Multiple Linear Regression')
plt.plot(GB_accuracy_df['DataFrame'],GB_accuracy_df['Accuracy'],label='Gradient Boosting Algorithm')
plt.xlabel('Data-Frames')
plt.ylabel('Accuracy')
plt.show()
```
